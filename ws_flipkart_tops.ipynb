```python
import requests                  # For making HTTP requests
from bs4 import BeautifulSoup     # For parsing HTML content
import pandas as pd  
```

```python
url="https://www.flipkart.com/search?q=tops+for+women&sid=clo%2Cash%2Cohw%2C36j&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=RECENT&suggestionId=tops+for+women%7CWomen%27s+Tops&requestId=995474bd-b526-4896-ab47-e9d335cd81ce&as-searchtext=tops%20for%20women"
```

```python
response=requests.get(url)
response
```

```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
```

```python
response=requests.get(url,headers=headers)
```

```python
response
```

```python
soup=BeautifulSoup(response.content,"html.parser")
```

```python
soup
```

```python
print(soup.prettify())
```

```python
products = soup.find_all("div", class_="hCKiGj")  
products
```

```python
len(products)
```

```python
product_top=[]
```

```python
for product in products:
    name = product.find("a",class_="WKTcLC")
    name = name.text if name else "N/A"

    price = product.find("div", class_="Nx9bqj")
    price = price.text if price else "N/A"
    product_top.append({
        'Title': name,
        'Price': price,
})

    
```

```python
print(product_top)
```

```python
df=pd.DataFrame(product_top)
```
